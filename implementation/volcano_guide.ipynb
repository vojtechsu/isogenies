## Isogeny graph

We can create a general isogeny graph from curve E with multiple degrees:

```python
from isogenies.volcano import Isogeny_graph
E = E = EllipticCurve(GF(17),[3,16])
G = Isogeny_graph(E,[3,5])
G.plot(vertex_size = (400,400))
```

```python
print(G.vertices())
print(G.edges())
```
